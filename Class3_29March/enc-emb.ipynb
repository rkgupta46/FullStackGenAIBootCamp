```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np
```

```python
print ("Hello, World!")
```

```python

documents = ["people watch movie", 
             "people watch cricket", 
             "people like movie", 
             "people like cricket"]
```

```python
token= [sentence.lower().split() for sentence in documents]
print(token)
```

```python
words_all = [[word] for sentence in token for word in sentence]
print (words_all)
```

```python
encoder = OneHotEncoder(sparse_output=False)
encoder.fit(words_all)
print (encoder.categories_[0])
```

```python
print (token)
for sentence in token:
    print (sentence)
```

```python
print (encoder.categories_[0])
for sentence in token:
    en_sen = encoder.transform([[word] for word in sentence])
    print ("\nsentence:", sentence)
    print (en_sen)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
documents = ["people watch movie and watch movie again", 
             "people watch cricket and watch cricket",
             "people like movie and like movie a lot",
             "people like cricket"]
```

```python
bow = CountVectorizer()
```

```python
bow_mat = bow.fit_transform(documents)
print (bow_mat.toarray())
print (bow.get_feature_names_out())
print (bow_mat.toarray().shape)
print (bow_mat.toarray()[0])
```
