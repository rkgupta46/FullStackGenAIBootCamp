```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python

documents = ["people watch movie and watch movie again", 
             "people watch cricket and watch cricket",
             "people like movie and like movie a lot",
             "people like cricket"]
```

```python
tf_documents = TfidfVectorizer()
tf_matrix = tf_documents.fit_transform(documents)
print(tf_matrix.toarray())
print(tf_documents.get_feature_names_out())
print(tf_matrix.shape)
```
